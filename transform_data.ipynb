```python
import os
import pandas as pd
```

```python
categories = pd.read_excel('data/processed/categories.xlsx')
categories
```

```python
df = pd.read_excel('data/processed/coded_data_2.xlsx')
df = df.drop(columns=['Unnamed: 0'])
df = df.drop(columns=['Unnamed: 0.1'])

# drop cases with no category
df = df.dropna(subset=['category'])
df = df.rename(columns={'category_id ': 'category_id'})

df = df.merge(categories, on='category_id', how='left')

df = df.rename(columns={
    'positve': 'positive',
    'column_name': 'Fragen Typ',
    'document': 'Dokument Nr.',
    'Ueberkategorie': '√úberkategorie',
    'assagekraft': 'Aussagekraft',
    'value': 'Text'})

# remove negative cases
df = df[df['positive'] == True]

df['Fragen Typ'] = df['Fragen Typ'].str.replace('_comment', '')
```

```python
df.columns
```

```python
info_df = df[['Dokument Nr.', 'Fragen Typ', '√úberkategorie', 'Kategorie', 'Aussagekraft', 'category_probability', 'rating', 'key_aspect', 'short_sumary', 'Text']]
anhang_df = df[['Dokument Nr.', 'Fragen Typ', '√úberkategorie', 'Kategorie', 'Text']]
```

```python
# Create a Pandas Excel writer using XlsxWriter or openpyxl as the engine
writer = pd.ExcelWriter('multiple_sheets.xlsx', engine='xlsxwriter')

# Write each DataFrame to a different worksheet
info_df = df[['Dokument Nr.', 'Fragen Typ', '√úberkategorie', 'Kategorie', 'Aussagekraft', 'category_probability', 'rating', 'key_aspect', 'short_sumary', 'Text']]
info_df.to_excel(writer, sheet_name='Info')

anhang_df = df[['Dokument Nr.', 'Fragen Typ', '√úberkategorie', 'Kategorie', 'Text']]
info_df.to_excel(writer, sheet_name='Anhang')

main_category_count = df.groupby(['√úberkategorie'])[['Dokument Nr.']].count().reset_index()
main_category_count = main_category_count.rename(columns={'Dokument Nr.': 'count'})
info_df.to_excel(writer, sheet_name='√úberkategorie')

category_count = df.groupby(['Kategorie'])[['Dokument Nr.']].count().reset_index()
category_count = category_count.rename(columns={'Dokument Nr.': 'count'})
info_df.to_excel(writer, sheet_name='Kategorie')

main_question_category_count = df.groupby(['√úberkategorie', 'Fragen Typ'])[['Dokument Nr.']].count().reset_index()
main_question_category_count = main_question_category_count.rename(columns={'Dokument Nr.': 'count'})
info_df.to_excel(writer, sheet_name='√úberkategorie-Fragen Typ')

category_question_count = df.groupby(['Kategorie', 'Fragen Typ'])[['Dokument Nr.']].count().reset_index()
category_question_count = category_question_count.rename(columns={'Dokument Nr.': 'count'})
info_df.to_excel(writer, sheet_name='Kategorie-Fragen Typ')

# Close the Pandas Excel writer and output the Excel file
writer.save()
```
